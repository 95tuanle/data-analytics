# Visualization Exercises

 Import Numpy, Panda, Matplotlib and any other necessary library.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Assign the following variables:

```python
x = np.arange(0,50)
y = x*2
z = x**2
print(x)
print(y)
print(z)
```

Draw y and z versus x in the same plot. The plot must be compelete (contain lables, legend, etc.)

```python
plt.plot(x, y, 'r', label='x vs y')
plt.plot(x, z, 'b', label='x vs z')
plt.xlabel('X Axis')
plt.ylabel('Y and Z Axes')
plt.title('Title')
plt.legend(loc="upper left")
plt.show()
```


Draw a two points at these coordinates: (0,0) and (5,130). The color of points must be green.

Read the titanic dataset as pandas dataframe

```python
df_titanic = pd.read_csv("Titanic.csv")
df_titanic.head()
```

Create a bar chart from the Age column of the dataset. 

```python
age = df_titanic.Age
age.plot(kind="bar")
```

Create a pie chart from the Age column of the dataset.  

```python
plt.pie(age.notnull())
```

Create a swarmplot of the Age and Sex of the dataset.

```python
sns.swarmplot(y = 'Age', x = 'Sex', data = df_titanic)
```

Plot heatmap of the dataset with corresponding annotations.


```python
corr_titanic = df_titanic.corr()
sns.heatmap(corr_titanic, annot = True)
```

Draw a scatter plot of the Age vs. Fare from the dataset. 

```python
plt.scatter(df_titanic.Age, df_titanic.Fare)
plt.xlabel('Age')
plt.ylabel('Fare')
```
